# COVID-19: Dados de Quantidades de Pessoas por Região

```python
# imports e inicializações

import matplotlib.pyplot as plt
import pandas as pd

#Colunas do arquivo Excel
#colunas = ["regiao", "casosNovos", "casosAcumulados", "obitosNovos", "obitosAcumulados"]
colunas = ["regiao", "casosNovos", "obitosNovos"]
regioes = ["Norte", "Nordeste", "Sudeste", "Sul", "Centro-Oeste"]

#Tamanho dos plots
plt.rcParams['figure.figsize'] = (11,7)

#leitura do excel para um dataframe
df = pd.read_excel('arquivo_geral.xlsx', usecols = colunas)

#Inicialização das listas principais dos plots
dados = []
casos = [1, 2]
```

## Região Norte

```python
#Criação de chart para região Norte

reg = regioes[0]
var = df.query('regiao == "' + reg + '"') 

for col in colunas[1:]: #Operação para cada modalidade
    dados.append(int(var[col].sum()))

# Plot por região
plt.title("Região " + reg + " (AC, AM, AP, PA, RO, RR, TO) - Janeiro-Abril de 2020")
plt.xlabel("Infectados/Óbitos")
plt.ylabel("Quantidade de pessoas")
plt.bar(casos, dados, width=0.3)
plt.xticks(casos, colunas[1:])

#Imprime os dados de cada sum
for i in range(len(dados)):
    plt.annotate(str(dados[i]), xy=(casos[i] - 0.1,dados[i]), xytext=(15, 3), 
                textcoords="offset points",
                ha='center', va='bottom')
    
plt.show()
dados[:] = []
        
```
